```python
## 1. Student Grade System
marks = int(input("Enter marks: "))

if marks < 35:
    print("Fail")
else:
    if marks < 75:
        print("Grade C")
    else:
        if marks < 90:
            print("Grade B")
        else:
            print("Grade A")
```

```python
## Number Check
num = int(input("Enter number: "))

if num % 2 == 0:
    if num % 4 == 0:
        print("Special Even")
    else:
        print("Even")
else:
    if num % 3 == 0:
        print("Special Odd")
    else:
        print("Odd")
```

```python
## . Age Classification
age = int(input("Enter age: "))

if age < 18:
    print("Minor")
else:
    if age < 60:
        print("Adult")
    else:
        print("Senior Citizen")
```

```python
## Temperature Category
temp = int(input("Enter temperature: "))

if temp < 15:
    print("Cold")
else:
    if temp <= 30:
        print("Moderate")
    else:
        if temp <= 40:
            print("Hot")
        else:
            print("Very Hot")
```

```python
## ATM Withdrawal
balance = int(input("Enter balance: "))
amount = int(input("Enter withdrawal amount: "))

if amount <= balance:
    if amount % 100 == 0:
        print("Transaction Successful")
    else:
        print("Amount must be multiple of 100")
else:
    print("Insufficient balance")
```

```python
## Login Validation
stored_user = "admin"
stored_pass = "1234"

user = input("Enter username: ")
pwd = input("Enter password: ")

if user == stored_user:
    if pwd == stored_pass:
        print("Login Successful")
    else:
        print("Incorrect password")
else:
    print("Invalid user")
```

```python
## Electricity Bill
units = int(input("Enter units: "))

if units <= 50:
    bill = units * 2
else:
    if units <= 100:
        bill = units * 3
    else:
        if units <= 200:
            bill = units * 5
        else:
            bill = units * 8

print("Bill amount:", bill)
```

```python
## Shopping Discount
amount = int(input("Enter amount: "))

if amount < 500:
    discount = 0
else:
    if amount < 1000:
        discount = amount * 0.05
    else:
        if amount < 5000:
            discount = amount * 0.10
        else:
            discount = amount * 0.20
print("discount:",discount)
print("Final amount:", amount - discount)
```

```python
## Loan Eligibility
salary = int(input("Enter salary: "))
credit = int(input("Enter credit score: "))

if salary >= 25000:
    if credit >= 750:
        print("Approved")
    else:
        print("Review")
else:
    if credit >= 800:
        print("Approved")
    else:
        print("Rejected")
```

```python
## Exam Result System
m1 = int(input("Enter mark1: "))
m2 = int(input("Enter mark2: "))
m3 = int(input("Enter mark3: "))

if m1 < 20 or m2 < 20 or m3 < 20:
    print(" severe Fail ")
else:
    if m1 >= 35 and m2 >= 35 and m3 >= 35:
        avg = (m1 + m2 + m3) / 3

        if avg >= 85:
            print("Distinction")
        else:
            if avg >= 60:
                print("First Class")
            else:
                print("Pass Class")
    else:
        print("Fail")
```

```python
## Ride Fare Calculation
distance = int(input("Enter distance: "))
peak = input("Peak time? (yes/no): ")

if distance <= 5:
    if peak == "yes":
        fare = distance * 15
    else:
        fare = distance * 10
else:
    if peak == "yes":
        fare = distance * 12
    else:
        fare = distance * 8

print("Fare:", fare)
```

```python
## Password Strength
pwd = input("Enter password: ")
length = len(pwd)

if length < 5:
    print("Very Weak")
else:
    if length < 8:
        print("Weak")
    else:
        if length < 12:
            print("Medium")
        else:
            print("Strong")
```

```python
## Employee Bonus
exp = int(input("Enter experience: "))
salary = int(input("Enter salary: "))

if exp < 2:
    print("No bonus")
else:
    if exp < 5:
        print("10% bonus")
    else:
        if salary < 50000:
            print("15% bonus")
        else:
            print("20% bonus")
```

```python
## Movie Ticket Pricing
age = int(input("Enter age: "))
show = input("Evening show? (yes/no): ")

if age < 12:
    price = 100
else:
    if age < 60:
        price = 200
    else:
        price = 150

if show == "yes":
    price += 50

print("Ticket price:", price)
```

```python
## Food Delivery Charges
order = int(input("Enter order amount: "))
distance = int(input("Enter distance: "))

if order < 500:
    charge = 50
else:
    charge = 0

if distance > 10:
    charge += 30

print("Delivery charge:", charge)
```

```python
## Employee Performance Rating
score = int(input("Enter score: "))
attendance = int(input("Enter attendance %: "))

if score < 50:
    print("Poor")
else:
    if score < 70:
        print("Average")
    else:
        if score < 90:
            print("Good")
        else:
            if attendance < 75:
                print("Good (Attendance issue)")
            else:
                print("Excellent")
```
