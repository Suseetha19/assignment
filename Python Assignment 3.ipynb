```python
# 1. Square Hollow Pattern

n = 5
for i in range(n):
    for j in range(n):
        if i == 0 or i == n-1 or j == 0 or j == n-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```

```python
# 2. Number Triangular

n = 4
for i in range(1, n+1):
    for j in range(i):
        print(i, end=" ")
    print()
```

```python
# 3. Number Increasing Pyramid

n = 4
for i in range(1, n+1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
# 4. Number Increasing Reverse Pyramid

n = 4
for i in range(n, 0, -1):
    for j in range(1, i+1):
        print(j, end=" ")
    print()
```

```python
# 5. Number Changing Pyramid

n = 4
num = 1

for i in range(1, n+1):
    for j in range(i):
        print(num, end=" ")
        num += 1
    print()
```

```python
# 6. Zero-One Triangle

n = 4

for i in range(1, n+1):
    for j in range(1, i+1):
        if (i + j) % 2 == 0:
            print(1, end=" ")
        else:
            print(0, end=" ")
    print()
```

```python
# 7. Palindrome Triangular

n = 4

for i in range(1, n+1):

    for j in range(i, 0, -1):
        print(j, end=" ")

    for j in range(2, i+1):
        print(j, end=" ")

    print()
```

```python
# 8. Rhombus Pattern

n = 5

for i in range(n):

    for j in range(n-i-1):
        print(" ", end=" ")

    for j in range(n):
        print("*", end=" ")

    print()
```

```python
# 9. Diamond Pattern

n = 5

for i in range(1, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(2*i-1):
        print("*", end=" ")

    print()

for i in range(n-1, 0, -1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(2*i-1):
        print("*", end=" ")

    print()
```

```python
# 10. Butterfly Star Pattern

n = 5

for i in range(1, n+1):

    for j in range(i):
        print("*", end=" ")

    for j in range(2*(n-i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()

for i in range(n-1, 0, -1):

    for j in range(i):
        print("*", end=" ")

    for j in range(2*(n-i)):
        print(" ", end=" ")

    for j in range(i):
        print("*", end=" ")

    print()
```

```python
# 19. Mirror Image Triangle Pattern

n = 4

for i in range(n, 0, -1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, i+1):
        print(j, end=" ")

    print()

for i in range(2, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, i+1):
        print(j, end=" ")

    print()
```

```python
# 20. Hollow Triangle Pattern

n = 5

for i in range(1, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1 or i == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# 21. Hollow Reverse Triangle Pattern

n = 5

for i in range(n, 0, -1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1 or i == n:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# 22. Hollow Diamond Pyramid

n = 5

for i in range(1, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()

for i in range(n-1, 0, -1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# 23. Hollow Hourglass Pattern

n = 5

for i in range(n, 0, -1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()

for i in range(2, n+1):

    for j in range(n-i):
        print(" ", end=" ")

    for j in range(1, 2*i):

        if j == 1 or j == 2*i-1:
            print("*", end=" ")
        else:
            print(" ", end=" ")

    print()
```

```python
# 24. Pascal's Triangle

n = 5

for i in range(n):

    num = 1

    for j in range(n-i-1):
        print(" ", end=" ")

    for j in range(i+1):
        print(num, end=" ")
        num = num * (i-j) // (j+1)

    print()
```

```python
# 25. Right Pascal's Triangle

n = 5

for i in range(1, n+1):
    for j in range(i):
        print("*", end=" ")
    print()

for i in range(n-1, 0, -1):
    for j in range(i):
        print("*", end=" ")
    print()
```
